```python
def find_median(df):
    cloth_mid = df[df['Category'] == 'Clothing']['Review Rating'].median()
    foot_mid = df[df['Category'] == 'Footwear']['Review Rating'].median()
    outerwear_mid = df[df['Category'] == 'Outerwear']['Review Rating'].median()
    Accessories_mid = df[df['Category'] == 'Accessories']['Review Rating'].median()
    return cloth_mid , foot_mid , outerwear_mid , Accessories_mid
```

```python
clot_mid , foot_mid , outer_mid , access_mid = find_median(df)
```

```python
df.isnull().sum()
```

```python
df[df['Category'] == 'Clothing']['Review Rating'].fillall(clot_mid)
df[df['Category'] == 'Footwear']['Review Rating'].fillna(foot_mid)
df[df['Category'] == 'Outerwear']['Review Rating'].fillna(outer_mid)
df[df['Category'] == 'Accessories']['Review Rating'].fillna(access_mid)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ',"_")
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
